```python
def load_hindi_stories(paths):
    all_text = ""
    for path in paths:
        with open(path, 'r', encoding='utf-8') as file:
            all_text += file.read().strip() + " "
    return all_text

raw_text = load_hindi_stories(file_paths)
print("Sample Text:", raw_text[:500])
```

```python
import re

def clean_hindi_text(text):
    text = re.sub(r'\s+', ' ', text)  # remove extra whitespace
    text = re.sub(r'[“”"]', '', text)  # remove special quotes
    text = re.sub(r'[^\u0900-\u097F\s।!?]', '', text)  # keep only Hindi chars and punctuations
    return text.strip()

cleaned_text = clean_hindi_text(raw_text)
print("Cleaned Sample:", cleaned_text[:500])
```

```python
import stanza

# Download the Hindi model if not already done
stanza.download('hi')
nlp = stanza.Pipeline('hi')

def extract_pos_words(sentences, target_pos):
    pos_words = set()
    for sent in sentences:
        doc = nlp(sent)
        for word in doc.iter_words():
            if word.upos in target_pos:
                pos_words.add(word.text)
    return pos_words
```

```python


def extract_all_pos_vocab(sentences, target_pos_tags):
    pos_vocab = {tag: set() for tag in target_pos_tags}

    for sent in sentences:
        doc = nlp(sent)
        for word in doc.iter_words():
            if word.upos in target_pos_tags:
                pos_vocab[word.upos].add(word.text)

    return {upos: list(words) for upos, words in pos_vocab.items()}
```

```python
def segment_with_pos_aware_rules(sentences, pronouns, conjunctions, connectives):
    segments = []
    current_segment = []

    for i, sentence in enumerate(sentences):
        first_word = sentence.split()[0] if sentence.strip() else ""

        # Check if first word is a pronoun or connective/conjunction — continue segment
        if first_word in pronouns or first_word in conjunctions or first_word in connectives:
            current_segment.append(sentence)
        else:
            # Check if previous sentence ended with a pronoun or context word
            if current_segment:
                segments.append(" ".join(current_segment))
            current_segment = [sentence]

    # Final append
    if current_segment:
        segments.append(" ".join(current_segment))

    return segments
```

```python
# Step 1: Get POS vocab
target_pos_tags = ['PRON', 'CCONJ', 'SCONJ']
pos_vocab = extract_all_pos_vocab(sentences, target_pos_tags)

# Step 2: Extract specific lists
pronouns = pos_vocab['PRON']
conjunctions = pos_vocab['CCONJ']
connectives = pos_vocab['SCONJ']

# Step 3: Segment
segments = segment_with_pos_aware_rules(sentences, pronouns, conjunctions, connectives)

# View segmented output
# for i, seg in enumerate(segments, 1):
#     print(f"[Segment {i}]: {seg}\n")
```

```python
import random
from itertools import combinations, product

def create_contrastive_pairs(segments, num_negative_samples_per_segment=2):
    positive_pairs = []
    negative_pairs = []

    # Positive pairs: from same segment
    for segment in segments:
        sentences = segment.strip().split("।")  # Hindi sentence delimiter
        sentences = [s.strip() for s in sentences if s.strip()]
        if len(sentences) >= 2:
            positive_pairs.extend(list(combinations(sentences, 2)))

    # Negative pairs: from different segments
    for i, seg1 in enumerate(segments):
        sents1 = seg1.strip().split("।")
        sents1 = [s.strip() for s in sents1 if s.strip()]
        if not sents1: continue

        for _ in range(num_negative_samples_per_segment):
            j = random.randint(0, len(segments) - 1)
            while j == i:
                j = random.randint(0, len(segments) - 1)
            sents2 = segments[j].strip().split("।")
            sents2 = [s.strip() for s in sents2 if s.strip()]
            if not sents2: continue

            # Randomly pick one sentence from each segment
            sent1 = random.choice(sents1)
            sent2 = random.choice(sents2)
            negative_pairs.append((sent1, sent2))

    return positive_pairs, negative_pairs
```

```python
positive_pairs, negative_pairs = create_contrastive_pairs(segments, num_negative_samples_per_segment=3)

print(" Sample Positive Pair:\n", positive_pairs[0])
print(" Sample Negative Pair:\n", negative_pairs[0])
print(f"\nTotal Positive Pairs: {len(positive_pairs)}")
print(f"Total Negative Pairs: {len(negative_pairs)}")
```

```python
from torch.utils.data import Dataset
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer

class ContrastivePairsDataset(Dataset):
    def __init__(self, positive_pairs, negative_pairs, model_name='sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2'):
        self.pairs = positive_pairs + negative_pairs
        self.labels = [1]*len(positive_pairs) + [0]*len(negative_pairs)

        self.tokenizer = AutoTokenizer.from_pretrained(model_name)

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        sent1, sent2 = self.pairs[idx]
        label = self.labels[idx]

        encoded = self.tokenizer(sent1, sent2, padding='max_length', truncation=True, return_tensors='pt', max_length=256)

        return {
            'input_ids': encoded['input_ids'].squeeze(0),
            'attention_mask': encoded['attention_mask'].squeeze(0),
            'label': label
        }
```

```python
from torch.utils.data import DataLoader, random_split

# Initialize dataset
dataset = ContrastivePairsDataset(positive_pairs, negative_pairs)

# Optionally split into train/val
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# Dataloaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
```

```python
import torch
import torch.nn as nn
from transformers import AutoModel

class SiameseContrastiveModel(nn.Module):
    def __init__(self, model_name='sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2'):
        super(SiameseContrastiveModel, self).__init__()
        self.encoder = AutoModel.from_pretrained(model_name)

    def forward(self, input_ids1, attention_mask1, input_ids2, attention_mask2):
        output1 = self.encoder(input_ids=input_ids1, attention_mask=attention_mask1)
        output2 = self.encoder(input_ids=input_ids2, attention_mask=attention_mask2)

        # Use [CLS] token embeddings
        emb1 = output1.last_hidden_state[:, 0, :]
        emb2 = output2.last_hidden_state[:, 0, :]

        return emb1, emb2
```

```python
from torch.nn import CosineEmbeddingLoss
import torch.optim as optim
from tqdm import tqdm

def train(model, train_loader, val_loader, epochs=5, lr=2e-5, device='cuda'):
    model.to(device)
    optimizer = optim.AdamW(model.parameters(), lr=lr)
    criterion = CosineEmbeddingLoss()

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for batch in tqdm(train_loader, desc=f"Training Epoch {epoch+1}"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            label = batch['label'].to(device)

            # Split into two sentence sets
            input_ids1 = input_ids[:, 0:128]
            attention_mask1 = attention_mask[:, 0:128]
            input_ids2 = input_ids[:, 128:]
            attention_mask2 = attention_mask[:, 128:]

            optimizer.zero_grad()
            emb1, emb2 = model(input_ids1, attention_mask1, input_ids2, attention_mask2)

            # Convert labels: 1 -> 1.0, 0 -> -1.0
            targets = 2 * label.float() - 1
            loss = criterion(emb1, emb2, targets)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        train_loss=total_loss/len(train_loader)
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for batch in tqdm(val_loader, desc=f"Validating Epoch {epoch+1}"):
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                label = batch['label'].to(device)

                # Split into two sentence sets for contrastive learning
                input_ids1 = input_ids[:, 0:128]
                attention_mask1 = attention_mask[:, 0:128]
                input_ids2 = input_ids[:, 128:]
                attention_mask2 = attention_mask[:, 128:]

                emb1, emb2 = model(input_ids1, attention_mask1, input_ids2, attention_mask2)

                # Convert labels: 1 -> 1.0, 0 -> -1.0
                targets = 2 * label.float() - 1
                val_loss = criterion(emb1, emb2, targets)
                total_val_loss += val_loss.item()

        val_loss = total_val_loss / len(val_loader)

        # Print training and validation loss for this epoch
        print(f"Epoch {epoch+1}/{epochs} — Train Loss: {train_loss:.4f} — Val Loss: {val_loss:.4f}")


```

```python
model = SiameseContrastiveModel()
```

```python


# Step 2: Start training
train(model, train_loader, val_loader, epochs=5, lr=2e-5, device='cuda')
```
